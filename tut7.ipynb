```python
df1 = pd.read_csv('/Users/indurangatheshan/Desktop/coursera/insurance_data.csv')
df1.head()
```

```python
plt.scatter(df1.age,df1.bought_insurance,marker='+',color='red')
```

```python
from sklearn.model_selection import train_test_split
x_test,x_train,y_test,y_train = train_test_split(df1[['age']],df1.bought_insurance ,test_size=0.1)
```

```python
from sklearn import linear_model
```

```python
model = linear_model.LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
