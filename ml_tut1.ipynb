```python
from sklearn import linear_model
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/indurangatheshan/Desktop/coursera/Book1.csv')
df
```

```python
%matplotlib inline
```

```python
plt.xlabel('area(sq ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price,color='green',marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3300]])
```
