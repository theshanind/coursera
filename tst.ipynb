```python
def chickenpox_by_sex():
    import pandas as pd
    
    df = pd.read_csv('/Users/indurangatheshan/Desktop/coursera/NISPUF17 (1).csv')
    
    vaccinated = df[df['P_NUMVRC']>=1]
    
    cpx_contracted = vaccinated[vaccinated['HAD_CPOX']==1]
    cpx_not_contracted = vaccinated[vaccinated['HAD_CPOX']==2]
    
    male_contracted = cpx_contracted[cpx_contracted['SEX']==1].shape[0]
    male_not_contracted = cpx_not_contracted[cpx_not_contracted['SEX']==1].shape[0]
    
    female_contracted = cpx_contracted[cpx_contracted['SEX']==2].shape[0]
    famale_not_contracted = cpx_not_contracted[cpx_not_contracted['SEX']==2].shape[0]
    
    male_ratio=male_contracted/male_not_contracted
    female_ratio=female_contracted/famale_not_contracted
    
    return {
        "male":male_ratio,
        "female":female_ratio
    }
print(chickenpox_by_sex())
```

```python
def avarage_influance_doses():
    import pandas as pd 
    
    df = pd.read_csv('/Users/indurangatheshan/Desktop/coursera/NISPUF17 (1).csv')
    
    breastfed = df[df['CBF_01']==1]
    not_breastfed = df[df['CBF_01']==2]
    
    avg_breastfed = breastfed['P_NUMFLU'].mean()
    avg_not_breastfed = not_breastfed['P_NUMFLU'].mean()
    
    return (avg_breastfed, avg_not_breastfed)
print(avarage_influance_doses())
```
